# Extracting LISTING data from json file








```python
#def listing():
path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

airbnb_data = []

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for i in data_list:
                    data = {
                        "Id": i.get('_id'),
                        "Listing_url": i.get('listing_url'),
                        "Name": i.get('name'),
                        "Description": i.get('description'),
                        "House_rules": i.get('house_rules'),
                        "Property_type": i.get('property_type'),
                        "Room_type": i.get('room_type'),
                        "Bed_type": i.get('bed_type'),
                        "Min_nights": int(i.get('minimum_nights')),
                        "Max_nights": int(i.get('maximum_nights')),
                        "Cancellation_policy": i.get('cancellation_policy'),
                        "Accomodates": i.get('accommodates'),
                        "Total_bedrooms": i.get('bedrooms'),
                        "Total_beds": i.get('beds'),
                        "Availability_365": i['availability'].get('availability_365'),
                        "Price": float(i.get('price')),
                        "Security_deposit": i.get('security_deposit'),
                        "Cleaning_fee": i.get('cleaning_fee'),
                        "Extra_people": float(i.get('extra_people')),
                        "Guests_included": float(i.get('guests_included')),
                        "No_of_reviews": i.get('number_of_reviews'),
                        "Review_scores": i['review_scores'].get('review_scores_rating'),
                        "Amenities": ', '.join(i.get('amenities', [])),
                        "Host_id": i['host'].get('host_id'),
                        "Host_name": i['host'].get('host_name'),
                        "Street": i['address'].get('street'),
                        "Country": i['address'].get('country'),
                        "Country_code": i['address'].get('country_code'),
                        "Location_type": i['address']['location'].get('type'),
                        "Longitude": i['address']['location']['coordinates'][0],
                        "Latitude": i['address']['location']['coordinates'][1],
                        "Is_location_exact": i['address']['location'].get('is_location_exact')
                    }
                    airbnb_data.append(data)
        except json.JSONDecodeError:
            print(f"Error reading file: {filepath}")

  #  return airbnb_data
# Now airbnb_data contains structured data from all JSON files in the directory.
```

```python
path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

amenities=[]

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for i in data_list:

                    Id=i.get("_id")
                    try:
                        amenity=i.get("amenities")
                    except:
                        amenity=None
                    amenities.append({"Id":Id,"amenity":amenity})
                aminitis=pd.DataFrame(amenities)
        except:
          print("error")
```

```python
import os
import json
import pandas as pd

path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

amenities_dict = {}

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for i in data_list:
                    Id = i.get("_id")
                    try:
                        amenity = i.get("amenities")
                    except:
                        amenity = None
                    if Id not in amenities_dict:
                        amenities_dict[Id] = []
                    if amenity:
                        amenities_dict[Id].extend(amenity)
        except Exception as e:
            print("Error:", e)

# Concatenate amenities for each Id into a single string
amenities_concatenated = {Id: ', '.join(amenities) if amenities else None for Id, amenities in amenities_dict.items()}

# Convert the dictionary to a DataFrame
amenities_df = pd.DataFrame.from_dict(amenities_concatenated, orient='index', columns=['amenities'])

# Reset index to make 'Id' a column instead of index
amenities_df.reset_index(inplace=True)
amenities_df.rename(columns={'index': 'Id'}, inplace=True)

amenities_df
```

# COLLECT HOST DETAILS

```python
import os
import json

path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

host_data = []

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for listing in data_list:
                    host_info = listing['host']
                    host_details = {
                        "Id": listing.get('_id'),
                        "Host_id": host_info.get('host_id'),
                        "Host_url": host_info.get('host_url'),
                        "Host_name": host_info.get('host_name'),
                        "Host_location": host_info.get('host_location'),
                        "Host_response_time": host_info.get('host_response_time'),
                        "Host_thumbnail_url": host_info.get('host_thumbnail_url'),
                        "Host_picture_url": host_info.get('host_picture_url'),
                        "Host_neighbourhood": host_info.get('host_neighbourhood'),
                        "Host_response_rate": host_info.get('host_response_rate'),
                        "Host_is_superhost": host_info.get('host_is_superhost'),
                        "Host_has_profile_pic": host_info.get('host_has_profile_pic'),
                        "Host_identity_verified": host_info.get('host_identity_verified'),
                        "Host_listings_count": host_info.get('host_listings_count'),
                        "Host_total_listings_count": host_info.get('host_total_listings_count'),
                        "Host_verifications": host_info.get('host_verifications')
                    }
                    host_data.append(host_details)
        except json.JSONDecodeError:
            print(f"Error reading file: {filepath}")
df_host=pd.DataFrame(host_data)
df_host

# Now 'host_data' contains host details for all listings
# You can process or analyze this data further as per your requirements

```

# COLLECT ADDRESS DETAILS

```python
import os
import json

path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

address_data = []

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for listing in data_list:
                    address_info = listing.get('address', {})
                    location_info = address_info.get('location', {})
                    address_details = {
                        "Id": listing.get('_id'),
                        "street": address_info.get('street'),
                        "suburb": address_info.get('suburb'),
                        "government_area": address_info.get('government_area'),
                        "market": address_info.get('market'),
                        "country": address_info.get('country'),
                        "country_code": address_info.get('country_code'),
                        "location_type": location_info.get('type'),
                        "longitude": location_info.get('coordinates', [])[0],
                        "latitude": location_info.get('coordinates', [])[1],
                        "is_location_exact": location_info.get('is_location_exact')
                    }
                    address_data.append(address_details)
        except json.JSONDecodeError:
            print(f"Error reading file: {filepath}")

# Create a DataFrame from the address data
df_address = pd.DataFrame(address_data)
df_address
```

```python
df_address.isnull().sum()
```

# COLLECT AVAILABILTY DETAILS

```python
import os
import json
import pandas as pd

path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

avail_data = []

for filename in filenames:
    filepath = os.path.join(path1, filename)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for listing in data_list:
                    _id = listing.get('_id')  # Corrected variable name
                    availability = listing.get("availability", {})
                    availability_30 = availability.get("availability_30")
                    availability_60 = availability.get("availability_60")
                    availability_90 = availability.get("availability_90")
                    availability_365 = availability.get("availability_365")

                    availability_details = {
                        "Id": _id,  # Changed from "Id" to "_id"
                        "availability_30": availability_30,
                        "availability_60": availability_60,
                        "availability_90": availability_90,
                        "availability_365": availability_365
                    }
                    avail_data.append(availability_details)
        except json.JSONDecodeError:
            print(f"Error reading file: {filepath}")

# Create a DataFrame from the availability data
df_availability = pd.DataFrame(avail_data)
df_availability
```

# COLLECT REVIEW SCORE DETAILS

```python
import os
import json
import pandas as pd

path1 = "drive/MyDrive/airbnb/"
filenames = os.listdir(path1)

revscore_data = []

for j in filenames:
    filepath = os.path.join(path1, j)
    with open(filepath, 'r', encoding='utf-8') as file:
        try:
            data_list = json.load(file)
            if isinstance(data_list, list):
                for listing in data_list:
                    revscore_info = listing.get('review_scores', {})
                    revscore_details = {
                        "Id": listing.get('_id'),
                        "Review_scores_accuracy": revscore_info.get("review_scores_accuracy"),
                        "Review_scores_cleanliness": revscore_info.get("review_scores_cleanliness"),
                        "Review_scores_checkin": revscore_info.get("review_scores_checkin"),
                        "Review_scores_communication": revscore_info.get("review_scores_communication"),
                        "Review_scores_location": revscore_info.get("review_scores_location"),
                        "Review_scores_value": revscore_info.get("review_scores_value"),
                        "Review_scores_rating": revscore_info.get("review_scores_rating")
                    }
                    revscore_data.append(revscore_details)
        except json.JSONDecodeError:
            print(f"Error reading file: {filepath}")

# Create a DataFrame from the review score data
df_revscore = pd.DataFrame(revscore_data)
df_revscore
```
